# 4. Carry out correlation analysis on the dataset and find the correlation factor between 
a. ph and Chloramines
b. ph and Conductivity
What do we infer from this correlation value?

```python
#correlation analysis on the dataset 
corrmatrix=df.corr()
plt.subplots(figsize=(20,8))
sns.heatmap(corrmatrix,vmin=.2,vmax=1,annot=True,linewidth=.2)
```

```python
#correlation between ph and chloramines
df1=df[['ph','Chloramines']]
corrmatrix1=df1.corr()
plt.subplots(figsize=(8,8))
sns.heatmap(corrmatrix1,vmin=.2,vmax=1,annot=True,linewidth=.2)
```

Corelation factor is -.025 ,which is approximately equal to zero.We can conclude that,there is no corelation between ph and Chloramines.

```python
#correlation between ph and conductivity
df2=df[['ph','Conductivity']]
corrmatrix2=df2.corr()
plt.subplots(figsize=(8,8))
sns.heatmap(corrmatrix2,vmin=.2,vmax=1,annot=True,linewidth=.2)
```

Corelation factor is .014 ,which is approximately equal to zero.We can conclude that,there is no corelation between ph and Conductivity.
